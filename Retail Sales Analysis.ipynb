## Section 4 : Data Quality Assessment, Cleaning and Transforming

```python
# Creating a copy before any Permanent Change of original dataset
df1 = df.copy()
```

```python
# Finding Missing Values Column-wise
df.isnull().sum()
```

```python
# As we don't know the Exact Postal Code and deleting entire Rows is not expeted, 
# So replacing Null values to integer Zero will be better for furthur calculation.
df1["Postal Code"].fillna(0, inplace=True)

df1["Postal Code"].isnull().sum()
```

```python
# Changing Postal Code column from Float to Integer
df1["Postal Code"].astype(int)
```

```python
# Duplicate Records
df1.drop_duplicates()

df1.duplicated().sum()
```

```python
# Outlier Detection (Sales & Profit)
sns.boxplot(x=df1["Sales"])
```

```python
sns.boxplot(x=df1["Profit"])
```

```python
# Date Formatting
df1["Order Date"] = pd.to_datetime(df1["Order Date"])
df1["Ship Date"] = pd.to_datetime(df1["Ship Date"])
```

```python
# Handle Invalid Values
df1 = df1[df1["Sales"]>0]
```

## Section 5 : Feature Engineering (Creating New Columns for KPI)

```python
# Creating a copy before any Permanent Change of original dataset
df2 = df1.copy()
```

```python
df2["Order Year"] = df2["Order Date"].dt.year
df2["Order Year"]
```

```python
df2["Order Quater"] = df2["Order Date"].dt.to_period("Q")
df2["Order Quater"]
```

```python
df2["Order Month"] = df2["Order Date"].dt.month
df2["Order Month"]
```

```python
df2["Delivery Time"] = (df2["Ship Date"]-df2["Order Date"]).dt.days
df2["Delivery Time"]
```

```python
# Handle Invalid Values
df2 = df2[df2["Delivery Time"]>0]
```

### New features created to support KPI and performance analysis:
- Profit Margin
- Shipping Delay
- Year, Quater and Month

## Step 9 : Time Series Analysis

```python
# Monthly Sales Trends
monthly_sales = df2.groupby(df2["Order Month"])["Sales"].sum()
print(monthly_sales)
monthly_sales.plot(
    kind = "line"
)
```

```python
# Year-Over-Year Growth
yearly_sales = df2.groupby(df2["Order Year"])["Sales"].sum()
print(yearly_sales)
yearly_sales.plot(
    kind = "line"
)
```

## Step 10 : Customer Segmentation Analysis

```python
# Top 20 Customers
top_cust = df2.groupby(df2["Customer Name"])["Sales"].sum().sort_values(ascending=False).head(20)
print(top_cust)
top_cust.plot(
    kind = "bar"
)
```

## Step 11 : Product Performance Analysis

```python
# Category level Performance
clp = df2.groupby("Category")[["Sales", "Profit"]].sum()
print(clp)
clp.plot(
    kind = "bar"
)
```

```python
# Sub-Category Loss Analysis
df2[df2["Profit"]<0].groupby("Sub-Category")["Profit"].sum().sort_values()
```

## Step 12 : Loss & Risk Analysis

```python
# Order with Negetive Profit Margin
df2["profit_margin"] = (df2["Profit"]/df2["Sales"])*100

df2[df2["profit_margin"]<0].sort_values(by="profit_margin").head(20)
```

```python
# Region-wise Loss
df2[df2["profit_margin"]<0].groupby("Region")["Profit"].sum()
```

## Section 13 : Advanced Interactive Visuals

```python
# Profit Margin by Category
fig = px.box(
    df2,
    x="Category",
    y="profit_margin",
    color="Category"
)
fig.show()
```

```python
# Average Profit Margin by Category
margin_cat = df2.groupby("Category", as_index=False)["profit_margin"].mean()

fig = px.bar(
    margin_cat,
    x="Category",
    y="profit_margin",
    color="profit_margin",
    text_auto=True,
    title="Average Profit Margin by Category"
)
fig.show()
```

```python
# Average Profit Margin by Region
margin_reg = df2.groupby("Region", as_index=False)["profit_margin"].mean()

fig = px.bar(
    margin_reg,
    x="Region",
    y="profit_margin",
    color="profit_margin",
    text_auto=True,
    title="Average Profit Margin by Region"
)
fig.show()
```

```python
# Profit Margin vs Discount
fig = px.scatter(
    df2,
    x="Discount",
    y="profit_margin",
    color="Category",
    title="Profit Margin vs Discount",
)
fig.show()
```

## Section 14 : Correlation Heatmap
